### Machete para el parcial de C3: Como Plotear en 3D

Hasta el momento, solo vimos como crear bellas gráficas de ploteos en 2 dimensiones, pero la vida real transcurre en 3D, y a veces nos gustaría visualizar estos fenómenos.<br>
Matplotlib nos permite crear gráficos en 3d con las mismas funciones que utilizabamos en 2d, si bien los argumentos son un poco distintos, si tienen este machete a mano no van a tener complicación alguna<br>

```python
import matplotlib.pyplot as plt
import numpy as np 

# por convensión, se suele llamar a los gráficos 3d como 'ax'

# para crear nuestro grafico y plotear sobre el distintos puntos en 3 dimensiones, debemos crear un 'subplot' con proyeccion en 3d y con los ejes en la misma escala
ax = plt.figure().add_subplot( projection='3d',aspect = 'equal'  )

# para luego, como haciamos en 2d,  plotear un scatterplot con los parametros 'x','y' y 'z': 
x = np.random.rand(200)* 2 - 1
y = np.random.rand(200)* 2 - 1
z = np.random.rand(200)* 2 - 1
ax.scatter(x,y,z, color = 'red') 
plt.show()
```

#### hora supongamos que queremos plotear vectores que simbolicen un campo vectorial.
Para ello, vamos a utiolizar la funcion 'quiver', que recibe como parámetros los puntos de la ubicacion, y el campo vectorial en si.



```python
#por ejemplo, supongamos un campo M = (-y/4,2x,z * -y)
u = -y/4
v = 2*x
w = z*(-y)
mag = np.linalg.norm([x,y,z])/75 #saco la magnitud para darle un largo a las flechitas, lo divido por 100 por que sino quedan enormes
ax = plt.figure().add_subplot( projection='3d',aspect = 'equal'  )
ax.quiver(x,y,z,u,v,w, length=mag, alpha = 0.7)
plt.show()
```
