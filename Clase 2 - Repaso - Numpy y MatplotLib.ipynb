#### MatplotLib
Es una librería estándar de python, es decir, viene incluída en la instalación del mismo y sirve para crear visualizaciones a partir de datos, pudiendo procesar gráficos estáticos, animados e interactivos en dos o trés dimensiones.</br>
Cuenta con una gran comunidad de desarollo opensource y vasta documentación, con ejemplos que bucear en su página oficial:</br>
https://matplotlib.org/</br>


Durante la cursada nos limitaremos a la función 'plot' del módulo 'pyplot', que dados puntos o arreglos de puntos (cosa que veremos mas abajo), produce gráficas bidimensionales en coordenadas rectangulares

```python
#sintaxis, grafico un punto

import matplotlib.pyplot as plt

def f(x):
    return x**2 - 5

for i in range(0,20):
    plt.plot(i,f(i),'.r')

plt.show()
# i  nuestro punto a evaluar
# f(i) su imagen
# '.r' indica que queremos graficar un punto, color rojo.
```

### Numpy
Es otra librería estándar de álgebra lineal para python, la principal forma de utilizar Numpy para la computación numérica es por medio de arreglos ( también llamados arrays o vectores ) como contenedores multidimensionales de información.</br>
Los arreglos son similares a las listas de python, la diferencia principal radica en que la lista puede contener datos de tipos distintos ( caracter, enteros, flotantes, booleanos, etc), contener sublistas en su interior y son unidimensionales; y el arreglo contiene tipos homogéneos de datos, pudiendo manejar operaciones aritméticas y trabajando en mas de una dimensión, es decir, un arreglo numpy puede contener multiples arreglos. 

```python
#Sintaxis
import numpy as np

#para crear un arreglo, puedo partir de una lista
datos_1 = [1,2,3]
arreglo = np.array(datos_1)
print(arreglo)

```

```python
#las operaciones con listas son bastante mas toscas, hay que recorrer índices uno por uno
#ejemplo, tengo una lista y un arreglo con los mismos valores
lista_1 = [1,10,20,24,45]
vector_1 = np.array([1,10,20,24,45])

def f(x):
    return x + (1/x)

#para la lista, tengo que recorrer cada uno de sus índices, creando una lista antes para que caigan los valores deseados
result_lista = []
for i in range(0, len(lista_1)) :
    result_lista.append(f(lista_1[i]))
print(result_lista)

#sin embargo, si trabajo con un arreglo de numpy, directamente puedo
result_vector = f(vector_1) 
print(result_vector)
```

```python
#yo puedo, a este arreglo, performarle operaciones básicas, como ser
datos_2 = np.sin(datos_1)
print('seno:') 
print(datos_2 )

datos_3 = np.cos(datos_1)
print('coseno:') 
print(datos_3)

#puedo sacar el maximo, el mínimo, la desviación, el promedio y demás operaciones
avg = np.mean(datos_1)
max = np.max(datos_1)
min = np.min(datos_1)
print(f'''promedio: {avg}, maximo: {max}, minimo: {min}''')

#puedo sacar la norma del vector
norm = np.linalg.norm(datos_1)
print('norma del vector:')
print(norm)

#al mismo tiempo, puedo realizar operaciones entre vectores como por ejemplo productos escalar y cruz
escalar = np.dot(datos_1,datos_2)
print(f'producto escalar: {escalar}')
vectorial = np.cross(datos_1,datos_2)
print('producto vectorial:')
print(vectorial)
```

```python
#para crear una matriz, puedo partir de dos listas
matriz = np.array([datos_1,datos_2,datos_2*np.sin(datos_1)])
print(matriz)

#y buscar su inversa!
inv = np.linalg.inv(matriz)
print('inversa de la matriz: ')
print(inv)
```

```python
#puedo crear vectores de números suscesivos
arreglo2 = np.linspace(0,5,50) #donde los parámetros son star, stop y cantidad de valores que quiero en mi arreglo
print('np.linspace:')
print(arreglo2)

arreglo3 = np.arange(0,5,0.1) #donde los parámetros son star, stop y paso entre estos
print('np.arange:')
print(arreglo3)

# o crear un vector vacío
empty = np.zeros(100) #donde el parámetro que recibe es la cantidad de indices
print('np.zeros:')
print(empty)
```

```python
#si quiero ver el valor de un registro en una matriz, voy a buscar por índices, que recordemos arrancan en cero
print('un solo valor, segunda fila segunda columa:') 
print(inv[1][1])
```

#### Volviendo a Matplotlib
La función plot en su forma mas básica puede recibir un solo argumento (un arreglo) que grafica como coordenada vertical, utilizando como coordenada horizontal el índice del vector


```python
test = np.sin(arreglo2)
plt.plot(test)
plt.show()
```

Pero la sintaxis habitual es introducir dos argumentos, donde el primero contiene un arreglo con los valores de la coordenada horizontal y el segundo en la coordenada vertical, a su vez, la función puede recibir distintas configuraciones como argumento para la visualización, color, estilo de linea y demases

```python
x = np.linspace(0,50,50)
y = x * np.sin(x)

plt.plot(x,y,'r--',label = 'partícula')
plt.xlabel('eje x')
plt.ylabel('imagen')
plt.title('plot prueba')
plt.legend()
plt.show()
```
